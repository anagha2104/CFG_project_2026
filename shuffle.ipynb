```python
# Read all the chromosome files. Based on input of protein name, shuffle the sequences from all files in 2 categories: U and B.
import os
import random
import pandas as pd
def read_chromosome_files(folder_path, protein_name):
 # file names are chr_i_200bp_bins.tsv
    chromosome_files = [f for f in os.listdir(folder_path) if f.startswith('chr') and f.endswith('_200bp_bins.tsv')]
    # we will add B and U sequences in different lists. And keep adding them for all files
    B_sequences = []
    U_sequences = []
    # so we need to take columns named "chr", "start", "end" for each row B and U respectively
    # we will read the file using pandas and filter the rows based on protein name and category
    # then we will add the sequences to respective lists
    # finally we will return the two lists
    # we will shuffle the sequences in each list before returning
    # we will use random.shuffle to shuffle the sequences in each list
    # There will be a column with the protein name. We fiter based on the value of that protein column for the row: B or U.
    for file in chromosome_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, sep='\t', header = 0)
        # filter rows based on protein name and category
        B_rows = df[df[protein_name] == 'B']
        U_rows = df[df[protein_name] == 'U']
        # add sequences to respective lists
        B_sequences.extend(B_rows[['chr', 'start', 'end']].values.tolist())
        U_sequences.extend(U_rows[['chr', 'start', 'end']].values.tolist())
    # shuffle the sequences in each list
    random.shuffle(B_sequences)
    random.shuffle(U_sequences)
    return B_sequences, U_sequences
# Call this function with the folder path and protein name to get the shuffled sequences for B and U categories
folder_path = '/content/drives/.shortcut-targets-by-id/1o3VAi95iwPeBb227O_GB8xueUaE_Ri9-/projectData' #'path_to_chromosome_files'
protein_name = 'CTCF' #'protein_of_interest'
B_sequences, U_sequences = read_chromosome_files(folder_path, protein_name)
# save the shuffled sequences as bed files
def save_as_bed_file(sequences, output_file):
    with open(output_file, 'w') as f:
        for seq in sequences:
            f.write('\t'.join(map(str, seq)) + '\n')
save_as_bed_file(B_sequences, 'B_sequences.bed')
save_as_bed_file(U_sequences, 'U_sequences.bed')


```
